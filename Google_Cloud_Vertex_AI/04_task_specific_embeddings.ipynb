Baed on the tasks, google supports different embedding models : https://cloud.google.com/vertex-ai/generative-ai/docs/embeddings/task-types
- Below are the task types :
1. semantic similarity
2. classification task
3. clustering
4. retrieval document / retrieval query / question answering / fact verification
5. code retrieval query

```python
import vertexai

# Initialize with your project and location
vertexai.init(project="your_project_name", location="us-central1")

from vertexai.language_models import TextEmbeddingModel, TextEmbeddingInput
model = TextEmbeddingModel.from_pretrained("text-embedding-004")
```

```python
text1 = TextEmbeddingInput("Why Sky is Blue?", "QUESTION_ANSWERING")
```

```python
text1_res = model.get_embeddings([text1])
```

```python
text1_res
```

```python
text1_res[0].values
```

```python
text2 = TextEmbeddingInput("Why Sky is Blue?", "CLUSTERING")
text2_res = model.get_embeddings([text2])
text2_res[0].values
```

We can see that for the same task, but different task type - the embeddings generated are different.
